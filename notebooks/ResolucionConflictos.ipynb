```python
import pandas as pd
```

```python
data = {
    'Producto': ['A', 'B', 'A', 'D'],
    'Ventas': [200, 150, 200, 250],
    'ciudad': ['Barcelona', 'Barcelona', 'Barcelona', 'Sevilla']
}
df= pd.DataFrame(data)
display(df)
```

```python
#Identificamos filas duplicadas
duplicados = df.duplicated()
print("Filas duplicadas:")
display(df[duplicados])
```

```python
#Eliminamos filas duplicadas
df_sin_duplicados = df.drop_duplicates()
display(df_sin_duplicados)
```

```python
#Resolver valores conflictivos
data_conflicto = {
    'Producto': ['A', 'B', 'C', 'D'],
    'Ventas_Fuente_1': [200, 150, 300, 250], 
    'Ventas_Fuente_2': [200, 145, 310, 250],
    'ciudad': ['Barcelona', 'Barcelona', 'Madrid', 'Sevilla']
}
df_conflicto = pd.DataFrame(data_conflicto)
print("DataFrame con valor conflictivo:")
display(df_conflicto)
```

```python
# Resolver conflicto eligiendo una fuente con el maximo
df_conflicto['Ventas_Resueltas'] = df_conflicto[['Ventas_Fuente_1', 'Ventas_Fuente_2']].max(axis=1)
display(df_conflicto)


```

```python
# Resolver conflicto eligiendo una fuente con la media
df_conflicto['Ventas_Resueltas'] = df_conflicto[['Ventas_Fuente_1', 'Ventas_Fuente_2']].mean(axis=1)
display(df_conflicto)
```

```python
datos_falta = {
    'Producto': ['A', 'B', 'C', 'D','E'],
    'Precio_1': [100, None, 200, 250,124],
    'Precio_2': [200, 150, None,122,350]
}

df_faltantes = pd.DataFrame(datos_falta)
print("DataFrame con valores faltantes:")
display(df_faltantes)
```

```python
# Rellenar los datos faltantes con la combinaci√≥n 
df_faltantes['Ventas_Resueltas'] = df_faltantes['Precio_1'].combine_first(df_faltantes['Precio_2'])
display(df_faltantes)
```

```python
#Media entre los dos precios con apply y una lambda
df_faltantes['Precio_Medio'] = df_faltantes[['Precio_1', 'Precio_2']].apply(lambda x: x.mean(), axis=1)
display(df_faltantes)
```

```python
#Interpolar los datos
df_faltantes['Precio_Interpolado'] = df_faltantes[['Precio_1','Precio_2']].interpolate(axis=1).iloc[:, 1]
display(df_faltantes)
```
