```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

## Escalamiento de Datos
El escalamiento de datos es una técnica para transformar los valores numéricos de un conjunto de datos a una misma escala o rango, sin distorsionar las diferencias entre ellos. Se usa principalmente en machine learning y análisis de datos para que todas las variables tengan el mismo peso y sean comparables.

Por ejemplo, si tienes una columna con valores entre 1 y 1000 y otra entre 0 y 1, el modelo puede dar más importancia a la de mayor rango. El escalamiento soluciona esto.

Métodos comunes:

Min-Max Scaling: lleva los valores a un rango entre 0 y 1.
Standard Scaling (Z-score): transforma los datos para que tengan media 0 y desviación estándar 1.
Esto mejora el rendimiento de muchos algoritmos y facilita la interpretación de los resultados.

## Normalizacion de datos

La normalización de datos es una técnica para transformar los valores numéricos de un conjunto de datos a un rango común, normalmente entre 0 y 1. El objetivo es que todas las variables tengan la misma escala y sean comparables, evitando que una variable con valores grandes domine a las demás.

El método más común es el Min-Max Scaling:

[ X_{norm} = \frac{X - X_{min}}{X_{max} - X_{min}} ]

Esto es útil en algoritmos de machine learning y análisis de datos donde la escala de las variables afecta el resultado.

```python
data = {
    'Producto': ['Manzada', 'Bannao', 'Naranja', 'Pera', 'Uva'],
    'Precio': [1000, 500, 1500, 1200, 2000],
    'Ventas': [50, 200, 150, 100, 300],
    'Descuento': [1, 2, 15, 5, 25]
}

df = pd.DataFrame(data)
display(df)
```

```python
# Tecnica 1: Min-Max Scaling
scaler = MinMaxScaler()
columnas_numericas = ['Precio', 'Ventas', 'Descuento']
df_escaladod = df.copy()
df_escaladod[columnas_numericas] = scaler.fit_transform(df[columnas_numericas])

display(df_escaladod)
```

```python
# Normalización de Datos con standardscaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_normalizado = df.copy()
df_normalizado[columnas_numericas] = scaler.fit_transform(df[columnas_numericas])
display(df_normalizado)
```

```python
# Visualización de Datos
import matplotlib.pyplot as plt
import seaborn as sns
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.histplot(df['Precio'], ax=axes[0], kde=True).set_title('Precio Original')
sns.histplot(df_escaladod['Precio'], ax=axes[1], kde=True).set_title('Precio Escalado (Min-Max)')
sns.histplot(df_normalizado['Precio'], ax=axes[2], kde=True).set_title('Precio Normalizado (StandardScaler)')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
df[columnas_numericas].plot(kind='box', ax=axes[0], title='Original')
df_escaladod[columnas_numericas].plot(kind='box', ax=axes[1], title='Escalado (Min-Max)')
df_normalizado[columnas_numericas].plot(kind='box', ax=axes[2], title='Normalizado (StandardScaler)')
plt.show()
```
