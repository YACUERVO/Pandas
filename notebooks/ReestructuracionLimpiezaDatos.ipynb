```python
import pandas as pd
import numpy as np
```

```python
data = {
    'ID': range(1, 21),
    'Nombre': [None, 'Maria', 'Pedro', 'Ana', 'Luis', 'Carlos', 'Maria', 'Pedro', 'Ana', 'Luis', 'Carlos', 'Maria', 'Pedro', 'Ana', 'Luis', 'Carlos', 'Maria', 'Pedro', 'Ana', 'Luis'],
    'Edad': [20, 21, 22, 23, None, 40, 26, 27, 28, 29, None, 31, 47, 33, 34, 48, 36, 37, 38, 39],
    'Salario': [1000, 2000, None, 4000, 5000, 6000, 7000, 8000, 9000, 10000, 11000, 12000, 13000, 14000, 15000, None, 17000, 18000, None, 20000],
    'Ciudad': ['Madrid', 'Barcelona', None, 'Sevilla', 'Malaga', 'Cordoba', 'Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Malaga', 'Cordoba', 'Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Malaga', 'Cordoba', 'Madrid', 'Barcelona'],
    'Fecha_Ingreso': ['2020-01-01', None, '2020-01-03', '2020-01-04', '2020-01-05', '2020-01-06', '2020-01-07', '2020-01-08', '2020-01-09', None, '2020-01-11', '2020-01-12', '2020-01-13', '2020-01-14', '2020-01-15', '2020-01-16', '2020-01-17', '2020-01-18', '2020-01-19', '2020-01-20']
}
df = pd.DataFrame(data)
display(df)

```

```python
#Buscar datos faltantes
#Datos faltantes por columna
df.isnull().sum()
```

```python
#Rellenar el salario con la media de los datos faltantes
#El argumento inplace=True en pandas significa que la operación se hace directamente sobre el DataFrame original, sin necesidad de reasignar el resultado a una nueva variable.
df['Salario'].fillna(df['Salario'].mean(),inplace=True)
display(df)
```

```python
#El argumento inplace=True en pandas significa que la operación se hace directamente sobre el DataFrame original, sin necesidad de reasignar el resultado a una nueva variable.
df['Salario']= df['Salario'].fillna(df['Salario'].mean())
#tambien se puede hacer de esta manera
df.fillna({'Salario': df['Salario'].mean()},inplace=True)
display(df)
```

```python
#Eliminar los datos de nombre y fecha de ingreso que no estan N/A
df.dropna(subset=['Nombre','Fecha_Ingreso','Ciudad'], inplace=True)
display(df)
```

```python
#Reestruturacion de un dataframe 
df['Rango_Edad'] = pd.cut(df['Edad'], bins= [20,30,40,50],labels=['20-30','30-40','40-50'])
df_grouped = df.groupby('Rango_Edad')['Salario'].mean()
display(df_grouped)
```
