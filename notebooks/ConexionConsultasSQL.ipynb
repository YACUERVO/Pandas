```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
empleados = {
    "EmpleadoID": [1, 2, 3, 4, 5],
    "Nombre": ["Ana Torres", "Carlos Gómez", "Laura Pérez", "Javier Ruiz", "María López"],
    "Departamento": ["Recursos Humanos", "Finanzas", "TI", "Marketing", "Recursos Humanos"],
    "Cargo": ["Analista HR", "Contador", "Desarrollador", "Community Manager", "Jefe HR"],
    "Salario": [2500, 3200, 4000, 2800, 4500],
    "Fecha_Contratacion": [
        "2020-01-15",
        "2019-07-23",
        "2021-03-10",
        "2022-11-01",
        "2018-05-30"
    ],
    "Estado": ["Activo", "Activo", "Activo", "Activo", "Inactivo"]
}

df = pd.DataFrame(empleados)

display(df)
```

```python
#Guardar la base de datos en sqlLite
engine = create_engine('sqlite:///empleados.db')
```

```python
df.to_sql('empleados', con=engine, if_exists='replace', index=False)
```

```python
#Leer los empleados
resuld_df= pd.read_sql('SELECT * FROM EMPLEADOS', con=engine)
display(resuld_df)
```

```python
query = '''
SELECT Nombre, Departamento, Salario
FROM EMPLEADOS WHERE Departamento = 'Recursos Humanos'

'''

resultado = pd.read_sql(query, con=engine)
display(resultado)
```

```python
query = '''
SELECT Nombre, Departamento, Salario
FROM EMPLEADOS WHERE Salario > 3200

'''

resultado = pd.read_sql(query, con=engine)
display(resultado)
```
