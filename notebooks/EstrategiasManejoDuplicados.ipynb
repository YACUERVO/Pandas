```python
import pandas as pd
```

```python
datos = [
    {"ID": 1, "nombre": "Ana", "edad": 28, "salario": 2500, "fecha_ingreso": "2020-01-15", "ciudad": "Bogotá"},
    {"ID": 2, "nombre": "Luis", "edad": 35, "salario": 3200, "fecha_ingreso": "2019-03-20", "ciudad": "Medellín"},
    {"ID": 3, "nombre": "María", "edad": 40, "salario": 4000, "fecha_ingreso": "2018-07-10", "ciudad": "Cali"},
    {"ID": 3, "nombre": "María", "edad": 40, "salario": 4000, "fecha_ingreso": "2018-07-10", "ciudad": "Cali"},
    {"ID": 4, "nombre": "Ana", "edad": 50, "salario": 5000, "fecha_ingreso": "2017-11-01", "ciudad": "Barranquilla"},
    {"ID": 5, "nombre": "Lucía", "edad": 30, "salario": 2800, "fecha_ingreso": "2021-05-25", "ciudad": "Tunja"},
    {"ID": 5, "nombre": "Lucía", "edad": 30, "salario": 2800, "fecha_ingreso": "2021-05-25", "ciudad": "Tunja"},
    {"ID": 6, "nombre": "Andrés", "edad": 45, "salario": 4500, "fecha_ingreso": "2016-09-13", "ciudad": "Cartagena"}
]

df = pd.DataFrame(datos)
display(df)
```

```python
#Eliminar datos de filas duplicadas
df_no_duplicated = df.drop_duplicates()
display(df_no_duplicated)
```

```python
#sumarizacion de duplicados para el salario 
#reset_index()
#Al usar groupby, el índice pasa a ser la columna agrupada (nombre).
#Con .reset_index() restauras el índice numérico normal y vuelves a tener nombre como una columna.
# .agg({...})
# Aquí defines qué operación quieres aplicar a cada columna dentro de cada grupo:

# 'edad': 'first' → toma el primer valor de la edad encontrado en cada grupo.
# (Si todas las filas de un mismo nombre tienen la misma edad, no hay problema. Pero si difieren, solo agarra la primera).

# 'salario': 'sum' → suma los salarios de todas las filas del mismo nombre.
# (Aquí es donde se "resumen" los duplicados en una sola fila con el salario total).

# 'fecha_ingreso': 'first' → toma la primera fecha de ingreso de cada grupo
df_sumarized = df.groupby('nombre').agg({
    'edad': 'first',
    'salario':'mean',
    'fecha_ingreso': 'first'
        
}).reset_index()

display(df_sumarized)
```

```python
#Etiquetar los duplicados en filas
df['Es_Duplicado'] = df.duplicated()
display(df)
```

```python
#Etiquetar los duplicados en columnas
df['Es_Duplicado'] = df.duplicated(subset=['nombre'])
display(df)
```

```python
df_nuevo= pd.DataFrame({
    'ID': [1,2,3],
    'Nombre': ['Ana','Luis', 'Gabriela'],
    'Salario': [51000,61000,68000],
    'Fecha_ingreso': ['2021-01-15','2021-06-15','2021-08-01']   
    
})

df_nuevo = pd.DataFrame(df_nuevo)
display(df_nuevo)
```

```python
#Concatener dos tablas
df_merged = pd.concat([df , df_nuevo]).drop_duplicates(subset=['ID'],keep='last')
display(df_merged)
```
