```python
import pandas as pd 
```

```python
data = {
    'producto': ['Manzanas', 'Naranjas', 'Platanos', 'Uvas','Peras'],
    'precio': [100, 80, 60, 120, 90],
    'stock': [100, 50, 30, 20,10]  }

df = pd.DataFrame(data)

df 
```

```python
def descuento(precio):
    return precio * 0.85
# Aplicar descuento a la columna 'precio'
```

```python
#Descuento aplicado a la columna 'precio'
df['precio_descuento'] = df['precio'].apply(descuento)
```

```python
df
```

```python
def mayusculas(cadena):
    return cadena.upper()
```

```python
#Aplicar la función mayusculas a la columna 'producto'
# Esto convertirá los nombres de los productos a mayúsculas
df['producto_mayusculas'] = df['producto'].apply(mayusculas)
df
```

```python
#Funcion lambda para convertir a mayúsculas
# Esto es una forma más concisa de aplicar la misma transformación
df['producto_mayusculas_lambda'] = df['producto'].apply(lambda x: x.upper())
df
```

```python
df['producto_map'] = df['producto'].map({'Manzanas' :'Apple', 'Naranjas' :'Orange'})
df
```

```python
#Aplicar aplymap para convertir a minúsculas
# Esto convertirá todos los valores de tipo cadena a minúsculas
df = df.applymap(lambda x: x.lower() if type(x) == str else x)
df
```

```python
#Aply para aplicar una función a lo largo de un eje del DataFrame
df['descuento'] = [10,15,10,20,5]
df
```

```python
df['precio_final'] = df.apply(lambda row: row['precio'] * (100 - row['descuento'])/100, axis=1)
df
```
