```python
import pandas as pd


data = {
    'nombre': ['Juan', 'Maria', 'Pedro', 'Ana', 'Luis'],
    'edad': [25, 30, 35, 40, 45],
    'ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Madrid'],
    'pais': ['España', 'España', 'España', 'España', 'España'],
    'telefono': ['1234567890', '0987654321', '1111111111', '2222222222', '3333333333'],
    'numero_hijos': [2, 1, 0, 3, 2],
    'casado': [True, False, True, False, True],
}

df = pd.DataFrame(data)
df
```

```python
#Seleccionar una columna
nombre = df['nombre']
nombre
```

```python
#Agregar una columna
df['salario'] = [50000, 60000, 70000, 80000, 90000]
df
```

```python
#Borrando una columna
df.drop('salario', axis=1, inplace=True)
df
```

```python
#Seleccionar una fila por posicion
primer_fila = df.iloc[4] #iloc es para seleccionar una fila por posicion
primer_fila
```

```python
#Seleccionar una fila por ciudad
df.loc[df['numero_hijos'] > 2]
```

```python
#Seleccionar por ciudad
df.loc[df['ciudad'] == 'Madrid']
```

```python
#Ordenar por edad de forma descendente
df.sort_values(by='edad', ascending=False)

```

```python
media_edad= df['edad'].mean()  # Media de la columna 'edad'
maxima_edad = df['edad'].max()   # Máximo de la columna 'edad'
minima_edad = df['edad'].min()   # Mínimo de la columna 'edad'
print(f"Media de edad: {media_edad}")
print(f"Máxima edad: {maxima_edad}")
print(f"Mínima edad: {minima_edad}")
```

```python
#Eliminar filas con NaN El parámetro inplace=True en df.dropna(inplace=True) significa que la operación se realiza directamente sobre el DataFrame original
df.dropna(inplace=True)
```

```python
df.fillna({'edad': 0, 'numero_hijos': 0}, inplace=True)  # Rellenar NaN con 0 en las columnas 'edad' y 'numero_hijos'
```
