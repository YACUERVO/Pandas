```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Producto': [
        'Manzana', 'Banana', 'Naranja', 'Naranja', 'Papaya',
        'Banana', 'Banana', 'Papaya', 'Naranja', 'Naranja',
        'Banana', 'Manzana', 'Manzana', 'Papaya', 'Banana'
    ],
    'Categoria': [
        'Fruta', 'Fruta', 'Tropical', 'Tropical', 'Tropical',
        'Fruta', 'Fruta', 'Tropical', 'Tropical', 'Tropical',
        'Fruta', 'Fruta', 'Fruta', 'Tropical', 'Fruta'
    ],
    'Ventas': [120, 95, 210, 175, 140, 200, 180, 130, 160, 145, 220, 155, 190, 170, 205],
    'Precio_Unitario': [1.20, 0.85, 1.50, 2.10, 1.75, 0.95, 1.10, 2.00, 1.65, 1.80, 1.05, 1.25, 1.40, 1.90, 1.00],
    'Fecha': pd.date_range(start='2025-09-01', periods=15, freq='D')
}

df = pd.DataFrame(data)
print("DataFrame Original:")
display(df)
```

```python
# Resumen de ventas por producto
resumen_ventas = df.groupby('Producto')['Ventas'].agg(['sum', 'mean', 'max']).reset_index()
resumen_ventas.columns = ['Producto', 'Ventas_Totales', 'Ventas_Promedio', 'Ventas_Maximas']
print("\nResumen de Ventas por Producto:")
display(resumen_ventas)
```

```python
# Calcular el rango de ventas por cada categoría
def rango_ventas(grupo):
    return grupo.max() - grupo.min()
rango_por_producto = df.groupby('Producto')['Ventas'].apply(rango_ventas).reset_index()
rango_por_producto.columns = ['Producto', 'Rango_Ventas']
print("\nRango de Ventas por Producto:")
display(rango_por_producto)
```

```python
# Transformacion en cada gurpo normalizar las ventas
df['Ventas_Normalizadas'] = df.groupby('Producto')['Ventas'].transform(lambda x: (x - x.mean()) / x.std())
print("\nDataFrame con Ventas Normalizadas:")
display(df[['Producto', 'Ventas', 'Ventas_Normalizadas']])
```

```python
# Agrupacion por multiples columnas
ventas_categoria = df.groupby(['Categoria', 'Producto'])['Ventas'].sum().reset_index()
ventas_categoria.columns = ['Categoria', 'Producto', 'Ventas_Totales']
print("\nVentas Totales por Categoria y Producto:")
display(ventas_categoria)
```

```python
# calcular el ingreso total por fila
# No, en este caso no necesitas usar axis=1.
# La expresión df['Ingreso_Total'] = df['Ventas'] * df['Precio_Unitario'] realiza una multiplicación elemento a elemento entre las columnas, es decir, multiplica cada fila de Ventas por la correspondiente de Precio_Unitario de forma automática.
# El parámetro axis=1 solo se usa en funciones como apply, sum, mean, etc., cuando quieres especificar que la operación se haga por filas.
# Aquí, la operación ya es fila a fila por defecto, así que no es necesario.


df['Ingreso_Total'] = df['Ventas'] * df['Precio_Unitario']
print("\nDataFrame con Ingreso Total:")
display(df)
```

```python
# Ingreso total por producto y categoria
df_ingreso_total = df.groupby(['Producto','Categoria']).agg({'Ingreso_Total':'sum'}).reset_index()
df_ingreso_total.columns = ['Producto', 'Categoria', 'Ingreso_Total']
display(df_ingreso_total)
```

```python
# Alternative way to do the same aggregation
df_ingreso_total = df.groupby(['Producto','Categoria']).agg(
    {'Ventas':'sum',
    'Ingreso_Total':'sum'}).reset_index()

display(df_ingreso_total)
```
