```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Empleado': ['Juan', 'Ana', 'Pedro', 'Luis','Maria','Ana','Pedro','Luis','Maria'],
    'Departamento': ['Ventas', 'Marketing','Ventas', 'IT','Marketing','IT', 'Ventas', 'Marketing','Ventas'],
    'Salario': [50000, 60000, 55000, 58000, 62000, 60000,55000,58000,62000 ],
    'Años_Experiencia': [5, 7, 6, 8, 9, 7, 6, 8, 9],
    'Ciudad': ['Madrid', 'Barcelona', 'Madrid', 'Sevilla', 'Barcelona', 'Sevilla', 'Madrid', 'Barcelona', 'Sevilla']})

display(df)

```

```python
#Obtener la suma o medio por cada departamento
df_agrupado = df.groupby('Departamento')['Salario'].sum()
print(df_agrupado)
```

```python
#Obtener la media de años de experiencia por departamento
media_experiencia = df.groupby('Departamento')['Años_Experiencia'].mean()
print(media_experiencia)
```

```python
#Agregación a una columnas
agregacion = df.groupby('Departamento')['Salario'].agg(['sum', 'mean', 'max', 'min'])
display(agregacion)
```

```python
#Agregación de múltiples columnas
agregacion = df.groupby('Departamento').agg({
    'Salario': ['sum', 'mean','max', 'min'],
    'Años_Experiencia': ['mean', 'max' , 'min']})
display(agregacion)

```

```python
#Agrupar por múltiples columnas y agregaciones complejas
agregacion_compleja = df.groupby(['Departamento', 'Empleado']).agg({
    'Salario': ['sum', 'mean'],
    'Años_Experiencia': ['mean'],
})
display(agregacion_compleja)
```
