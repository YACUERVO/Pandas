```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/Admin/Documents/proyectosAnalisisDatos/Pandas/datos_con_faltantes.csv', delimiter=',')
display(df.head())
```

```python
df['Ciudad']= df['Ciudad'].fillna('No registra')
df
```

```python
#Medidas de tendencia central 
#Medias
#Mediana
#Moda

#Datos faltantes de salario los llenaremos con la media
df['Salario']= df['Salario'].fillna(df['Salario'].mean())
df
```

```python
#Datos faltantes de la edad llenarlo con la mediana
#La mediana es bueno llenarlo para los datos atipicos
df['Edad']=df['Edad'].fillna(df['Edad'].median())
df
```

```python
df = pd.read_csv('/Users/Admin/Documents/proyectosAnalisisDatos/Pandas/datos_con_faltantes.csv', delimiter=',')
display(df)
```

```python
#El que no tiene ciudad el que no registra la ciudad corresponda la mayor cantidad de usuauro
df['Ciudad'] = df['Ciudad'].fillna(df['Ciudad'].mode()[0])
df
```

```python
df['Nombre'] = df['Nombre'].fillna('Desconocido')
df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Vamos a crear un modelo de regresion lineal en salario y edad para los valores faltantes
model = LinearRegression()
model.fit(df[['Edad']].dropna(), df['Salario'].dropna())
#Utulizar un aply para predecir los valores del salario
#Precedir el salario
df['Salario'] = df.apply(lambda row:model.predict([[row['Edad']]])[0] if pd.isnull(row['Salario']) else row ['Salario'], axis=1 )
df 
```

```python
from fancyimpute import IterativeImputer
from sklearn.experimental import enable_iterative_imputer

num_df = df.select_dtypes(include=['float64', 'int64'])

df_imputed = pd.DataFrame(
    IterativeImputer().fit_transform(num_df),
    columns=num_df.columns,
    index=num_df.index
)

df_imputed
```
