```python
import pandas as pd
import numpy as np  
from scipy import stats
```

```python
data = {
    'Nombre': ['Ana', 'Luis', 'Pedro', 'María'],
    'Edad': [23, 45, 34, 29]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
#Raiz cuadrada de la columna 'Edad'
df['sqrt_Age'] = np.sqrt(df['Edad'])
```

```python
df
```

```python
#Suma acumulativa de la columna 'Edad'
df['consum_B']= np.cumsum(df['Edad'])
```

```python
df
```

```python
data = {
    'x':np.random.rand(100),
    'y':np.random.rand(100) 
}
df = pd.DataFrame(data)
df
```

```python
#Correlación de Pearson
pearson_corr, p_value = stats.pearsonr(df['x'], df['y'])
print(f"Pearson Correlation: {pearson_corr}, p-value: {p_value}")
```

```python
#Regresión lineal
slope, intercept, r_value, p_value, std_err = stats.linregress(df['x'], df['y'])
print(f"Slope: {slope}, Intercept: {intercept}, R-squared: {r_value**2}, p-value: {p_value}, Std Err: {std_err}")
```

```python
#Sype avanzad
muestra = np.random.randn(1000) + 2
t_stat, p_value = stats.ttest_1samp(muestra, 2)
print(f"T-statistic: {t_stat}, p-value: {p_value}")
```
