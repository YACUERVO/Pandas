```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Empleado': ['Juan', 'Ana', 'Pedro', 'Luis','Maria','Ana','Pedro','Luis','Maria'],
    'Departamento': ['Ventas', 'Marketing','Ventas', 'IT','Marketing','IT', 'Ventas', 'Marketing','Ventas'],
    'Salario': [50000, 60000, 55000, 58000, 62000, 60000,55000,58000,62000 ],
    'A침os_Experiencia': [5, 7, 6, 8, 9, 7, 6, 8, 9],
    'Ciudad': ['Madrid', 'Barcelona', 'Madrid', 'Sevilla', 'Barcelona', 'Sevilla', 'Madrid', 'Barcelona', 'Sevilla']})

display(df)
```

```python
#Tabla dinamica con departamento donde sume los salarios
pivot_table = df.pivot_table(index='Departamento', values='Salario', aggfunc='sum')
display(pivot_table)
```

```python
#Varias funciones de agregaci칩n
pivot_table = df.pivot_table(index='Departamento', values='Salario', aggfunc=['sum','mean','max','min'])
display(pivot_table)
```

```python
#Tabla dinamica con multiples 칤ndices y columnas
pivot_table = df.pivot_table(index=['Departamento','Empleado'], columns='Ciudad', values='Salario', aggfunc='sum')
display(pivot_table)
```

```python
#Croosstab para contar empleados por departamento y ciudad
crosstab = pd.crosstab(df['Departamento'], df['Ciudad'])
display(crosstab)
```

```python
#Tables de contingencia con funciones de agregaci칩n
contingency_table = pd.crosstab(df['Departamento'], df['Ciudad'], values=df['Salario'], aggfunc='mean', margins=True)
display(contingency_table)
```
