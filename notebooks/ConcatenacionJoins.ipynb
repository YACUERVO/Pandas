```python
df_empleados = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Nombre': ['Juan', 'Ana', 'Pedro', 'Maria'],
    'Departamento': ['Ventas', 'Marketing', 'IT', 'Recursos Humanos']
})

# DataFrame de departamentos
df_departamento = pd.DataFrame({
    'Departamento': ['Ventas', 'Marketing', 'IT', 'Analista de datos'],
    'Jefe': ['Carlos', 'Sofía', 'Miguel', 'Laura']
})
```

```python
# Unión de DataFrames utilizando merge donde 'Departamento' es la clave en comun
# Utilizando 'inner' para mantener solo las filas con coincidencias en ambos DataFrames
df_inner = pd.merge(df_empleados, df_departamento, on='Departamento', how='inner')
display(df_inner)
```

```python
# Unión de DataFrames utilizando merge donde 'Departamento' es la clave en comun
# Utilizando left join para mantener todas las filas del DataFrame izquierdo
# y las coincidencias del DataFrame derecho
df_left_join = pd.merge(df_empleados, df_departamento, on='Departamento', how='left')
display(df_left_join)
```

```python
# Unión de DataFrames utilizando merge donde 'Departamento' es la clave en comun
# Utilizando right join para mantener todas las filas del DataFrame derecho
df_right_join = pd.merge(df_empleados, df_departamento, on='Departamento', how='right')
display(df_right_join)
```

```python
#outer join para mantener todas las filas de ambos DataFrames
df_outer_join = pd.merge(df_empleados, df_departamento, on='Departamento', how='outer')
display(df_outer_join)
```
