```python
import pandas as pd
from sqlalchemy import create_engine
    
from sqlalchemy import text
```

```python
engine = create_engine('sqlite:///:memory:')  # Usando una base de datos en memoria para el ejemplo 
```

```python
df = pd.DataFrame({
    'Empleado': ['Juan', 'Ana', 'Pedro', 'Luis','Maria','Ana','Pedro','Luis','Maria'],
    'Departamento': ['Ventas', 'Marketing','Ventas', 'IT','Marketing','IT', 'Ventas', 'Marketing','Ventas'],
    'Salario': [50000, 60000, 55000, 58000, 62000, 60000,55000,58000,62000 ],
    'Años_Experiencia': [5, 7, 6, 8, 9, 7, 6, 8, 9],
    'Ciudad': ['Madrid', 'Barcelona', 'Madrid', 'Sevilla', 'Barcelona', 'Sevilla', 'Madrid', 'Barcelona', 'Sevilla']})

display(df)
```

```python
#Guardar el DataFrame en la base de datos
df.to_sql('empleados', engine, index=False, if_exists='replace')
```

```python
#Leer datos desde la base de datos
df_db = pd.read_sql('SELECT * FROM empleados', engine)
display(df_db)
```

```python
df_db.info()
```

```python
# Consultas SQL
consultas_sql = "Select Empleado, Salario FROM empleados WHERE Departamento = 'Ventas' AND Años_Experiencia > 5"
df_consulta = pd.read_sql(consultas_sql, engine)
display(df_consulta)
```

```python
# Actualizar datos en la base de datos
with engine.connect() as connection:
    connection.execute(text("UPDATE empleados SET Salario = Salario * 1.1 WHERE Departamento = 'Ventas'"))
```

```python
#Verificar la actualización
df_actualizado = pd.read_sql('SELECT Departamento, Salario FROM empleados WHERE Departamento = "Ventas"', engine)
df_actualizado
```

```python
#Eliminar datos de la base de datos
with engine.connect() as connection:
    connection.execute(text("DELETE FROM empleados WHERE Departamento = 'Marketing'"))
```

```python
#Verificar la actualización
df_actualizado = pd.read_sql('SELECT * FROM empleados', engine)

display(df_actualizado)
```
