```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Empleado': ['Juan', 'Ana', 'Pedro', 'Luis','Maria','Ana','Pedro','Luis','Maria'],
    'Departamento': ['Ventas', 'Marketing','Ventas', 'IT','Marketing','IT', 'Ventas', 'Marketing','Ventas'],
    'Salario': [50000, 60000, 55000, 58000, 62000, 60000,55000,58000,62000 ],
    'Años_Experiencia': [5, 7, 6, 8, 9, 7, 6, 8, 9],
    'Ciudad': ['Madrid', 'Barcelona', 'Madrid', 'Sevilla', 'Barcelona', 'Sevilla', 'Madrid', 'Barcelona', 'Sevilla']})

display(df)
```

```python
#Normalizar los salarios dentro de cada departamento
df['Salario_Normalizado'] = df.groupby('Departamento')['Salario'].transform(lambda x: (x - x.mean()) / x.std())
display(df)
```

```python
#Filtrar por grupo
df_filtrado = df.groupby('Departamento').filter(lambda x: len(x) > 2)
display(df_filtrado)
```

```python
#Filtro .agg para obtener estadísticas
df_agg = df.groupby('Departamento').agg({
    'Salario': ['sum'],   
    'Años_Experiencia': ['mean']
})
display(df_agg)
```

```python
#Metodo Grouper para agrupar por intervalos de tiempo
df['Fecha'] = pd.to_datetime(['2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01', '2023-05-01', '2023-06-01', '2023-07-01', '2023-08-01', '2023-09-01'])
df['Mes'] = df['Fecha'].dt.to_period('M')
df['Ventas'] = [1000, 1500, 1200, 1300, 1600, 1400, 1100, 1700, 1800]

df  
```

```python
df.set_index('Fecha', inplace=True)
display(df)
```

```python
ventas_mensuales = df.groupby(pd.Grouper(freq='M')).sum()
display(ventas_mensuales)
```

```python
def rango_salario(grupo):
    return grupo['Salario'].max() - grupo['Salario'].min()
```

```python
rango_salario_result = df.groupby('Departamento').apply(rango_salario)
display(rango_salario_result)
```
