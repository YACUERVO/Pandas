```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/Admin/Documents/proyectosAnalisisDatos/Pandas/datos_con_faltantes.csv', delimiter=',')
display(df.head())
```

```python
#Primera opcion eliminar datos faltantes en filas
df_dropna_rows = df.dropna()
display(df_dropna_rows)

```

```python
#Segunda opcion eliminar columnas
df_dropna_cols = df.dropna(axis=1)
display(df_dropna_cols)
```

```python
#Tercer opcion rellenar los datos con datos constantes
df_fill_constant = df.fillna('Desconocido')
display(df_fill_constant)
```

```python
#Cuarta opcion rellenar columnas 
df_fill_name = df['Nombre'].fillna('Desconocido')
display(df_fill_name)
```
