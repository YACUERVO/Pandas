```python
import pandas as pd
```

```python
frutas = {
    'Producto': ['manzana', 'banana', 'naranja', 'pera', 'uva', 'sandía', 'kiwi', 'mango'],
    'Precio': [120, 180, 100, 150, 200, 350, 220, 280]
}

df = pd.DataFrame(frutas)
df
```

```python
#Establecer 'Producto'como índice   
df.set_index('Producto', inplace=True)
df
```

```python
#Reindexar el DataFrame con un nuevo orden de productos
#y agregar un nuevo producto 'Melones' al final
nuevo_orden = ['mango', 'kiwi', 'sandía', 'uva', 'pera', 'naranja', 'banana', 'manzana','melones']
df = df.reindex(nuevo_orden)
df
```

```python
#Renombrar los indices  
df_renombrar = df.rename(index={'Melones': 'melón'})
df_renombrar
```

```python
#Resetear el índice del DataFrame
df_reset = df.reset_index()
df_reset
```
