```python
import pandas as pd
```

```python
# Separar columnas con un delimitador
data = {
    'Nombre_Completo': ['Juan Perez', 'Ana Gomez', 'Luis Martinez']
}

df = pd.DataFrame(data)
display(df)
```

```python
# Separar la columna en dos con split
df[['Nombre', 'Apellido']] = df['Nombre_Completo'].str.split(' ', expand=True)
display(df)
#df = df.drop(columns=['Nombre_Completo'])
```

```python
# Separar columnas con varios delimitadores
data2 = {
    'Fecha_Hora': ['2023-10-01 14:30', '2023-10-02 09:15', '2023-10-03 18:45'] 
}

df2 = pd.DataFrame(data2)
display(df2)
```

```python
# Separar por dia, mes, año, hora y minuto
df2[['Fecha', 'Hora']] = df2['Fecha_Hora'].str.split(' ', expand=True)
df2[['Año', 'Mes', 'Día']] = df2['Fecha'].str.split('-', expand=True)
df2[['Hora', 'Minuto']] = df2['Hora'].str.split(':', expand=True)
display(df2)
```

```python
# Funcionar columnas o varias columnas
data3 = {
    'Nombre': ['Juan', 'Ana', 'Luis'],
    'Apellido': ['Perez', 'Gomez', 'Martinez']
}
df3 = pd.DataFrame(data3)
display(df3)

```

```python
# Combinar las columnas en una sola
df3['Nombre_Completo'] = df3['Nombre'] + ' ' + df3['Apellido']
display(df3)
```

```python
# Funcionar columnas con un formato especifico
data4 = {
    'Día': ['01', '02', '03'],
    'Mes': ['10', '10', '10'],
    'Año': ['2023', '2023', '2023']
}
df4 = pd.DataFrame(data4)
display(df4)

```

```python
# Combinar las columnas en una sola
df4['Fecha_Completa'] = df4[['Día', 'Mes', 'Año']].agg('-'.join, axis=1)
display(df4)
```

```python

# Ejemplo de datos con texto y números mezclados

# ¿Cómo funciona?

# str.extract(r'([A-Za-z]+)-(\d{4})') busca en cada valor de la columna:
# ([A-Za-z]+) → grupo de letras (el nombre)
# - → un guion
# (\d{4}) → grupo de 4 dígitos (el año)
data = {
    'Info': ['ProductoA-2023', 'ProductoB-2024', 'ProductoC-2025']
}
df = pd.DataFrame(data)

# Extraer el nombre y el año usando str.extract y una expresión regular
df[['Nombre', 'Año']] = df['Info'].str.extract(r'([A-Za-z]+)-(\d{4})')

display(df)
```

```python
data = {
        'Direccion': [
            'Calle 123 #45-67, Bogotá',
            'Carrera 10 #20-30, Medellín',
            'Avenida 5 #100-200, Cali',
            'Transversal 8 #50-60, Barranquilla'
           ]
           
}

df_dirreciones = pd.DataFrame(data)
display(df_dirreciones)
```

```python
#Extraer la calle, el número y la ciudad usando str.extract y una expresión regular
df_dirreciones[['Calle', 'Numero', 'Ciudad']] = df_dirreciones['Direccion'].str.extract(r'([A-Za-z\s]+)\s(\d+\s#\d+-\d+),\s([A-Za-z]+)')
display(df_dirreciones)
```

```python
data = {
    'Texto': [
        'Correo: juan123@gmail.com',
        'Teléfono: +57-300-1234567',
        'Fecha: 2025-09-08',
        'Precio: $150.50',
        'Código: ABC-1234'
    ]
}
df = pd.DataFrame(data)
display(df)

# 1. Extraer correo electrónico
df['Correo'] = df['Texto'].str.extract(r'([\w\.-]+@[\w\.-]+)')

# 2. Extraer número de teléfono internacional
df['Telefono'] = df['Texto'].str.extract(r'(\+\d{2}-\d{3}-\d{7})')

# 3. Extraer fecha en formato AAAA-MM-DD
df['Fecha'] = df['Texto'].str.extract(r'(\d{4}-\d{2}-\d{2})')

# 4. Extraer precio en formato $n.nn
df['Precio'] = df['Texto'].str.extract(r'(\$\d+\.\d{2})')

# 5. Extraer código tipo ABC-1234
df['Codigo'] = df['Texto'].str.extract(r'([A-Z]{3}-\d{4})')

display(df)
```

```python
# Ejemplo avanzado: extraer usuario, dominio y extensión de correos electrónicos


data = {
    'Emails': [
        'juan.perez@gmail.com',
        'ana_gomez123@outlook.com',
        'luis-martinez@empresa.edu.co',
        'soporte@mi-web.net'
    ]
}
df = pd.DataFrame(data)
display(df)

# Expresión regular corregida (solo una barra invertida y extensión con puntos)
df[['Usuario', 'Dominio', 'Extension']] = df['Emails'].str.extract(
    r'(?P<Usuario>[\w\.-]+)@(?P<Dominio>[\w\.-]+)\.(?P<Extension>[a-z\.]+)'
)

display(df)
```
