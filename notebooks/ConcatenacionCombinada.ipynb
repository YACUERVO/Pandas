```python
import pandas as pd
```

Concatenacion:
Es unir dataFrame ya sea aplicandolos verticalmente a lo largo de filas o combinacion horizontalmente a lo largo de las columnas 

Combinacion:
Funcionar DataFrame utilizando una columna clave o indice compartido 

```python
data_lima = {
    'Producto': ['A', 'B', 'C'],
    'Ventas': [100, 200, 300],
    'Ciudad': ['Lima', 'Lima', 'Lima']
}

data_bogota = {
    'Producto': ['D', 'E', 'C'],
    'Ventas': [150, 250, 350],
    'Ciudad': ['Bogota', 'Bogota', 'Bogota']    
}

df_lima = pd.DataFrame(data_lima)
df_bogota = pd.DataFrame(data_bogota)

display(df_lima)
display(df_bogota)
```

```python
#Concatener de forma vertical
df_concat = pd.concat([df_lima, df_bogota] , ignore_index=True)
display(df_concat)
```

```python
#Concatenacion horizontal
data_inventario = {
    'Producto': ['A', 'B', 'C', 'D', 'E'],
    'Inventario': [50, 60, 70, 80, 90]
}
df_inventario = pd.DataFrame(data_inventario)
display(df_inventario)


```

```python
#Concatenacion horizontal
# pd.concat([...], axis=1): concatena (une) los DataFrames que pongas en la lista, pero de forma horizontal, es decir, uno al lado del otro (por columnas).
# reset_index(drop=True): reinicia el Ã­ndice de cada DataFrame para que ambos tengan Ã­ndices que van de 0 a n-1, evitando problemas si los Ã­ndices originales no coinciden.
# [df_concat.reset_index(drop=True), df_inventario.reset_index(drop=True)]: se pasan ambos DataFrames con sus Ã­ndices reiniciados.
# El resultado es un nuevo DataFrame (df_h_concat) que tiene todas las columnas de ambos DataFrames, alineadas por el mismo Ã­ndice.
df_h_concat = pd.concat([df_concat.reset_index(drop=True), df_inventario.reset_index(drop=True)], axis=1)
display(df_h_concat)
```

```python
#Ejemplos con datos simples para explicar todas las combinaciones con merge
#Funcion merge para combinar DataFrame con una columna clave
df_concat = pd.DataFrame({
    "Producto": ["A", "B", "C", "D"],
    "Ventas": [100, 200, 300, 400]
})

df_inventario = pd.DataFrame({
    "Producto": ["A", "B", "E"],
    "Stock": [50, 80, 120]
})
```

```python
#Funcion merge para combinar DataFrame con una columna clave
# 1. inner (intersecciÃ³n)
# Devuelve solo los registros comunes en ambas tablas (intersecciÃ³n de Producto).
print('''Solo aparecen A y B porque son los Ãºnicos productos que existen en ambos DataFrames.
      inner : Solo productos comunes (A, B)''')
df_combinado = pd.merge(df_concat, df_inventario, on='Producto', how='inner')
display(df_combinado)

```

```python
# 2. left (todos de la izquierda + coincidencias de la derecha)
# Devuelve todos los registros del DataFrame izquierdo (df_concat) y agrega los datos de la derecha (df_inventario) si coinciden.
# Si no encuentra coincidencia â†’ coloca NaN.

print(''' Se conservan todos los productos de df_concat.
ðŸ‘‰ C y D no tienen stock, por eso aparece NaN.
left: Todos de ventas (A, B, C, D) + coincidencias''')

df_combinado_left=pd.merge(df_concat, df_inventario, on="Producto", how="left")
display(df_combinado_left)
```

```python
# 3. right (todos de la derecha + coincidencias de la izquierda)
# Devuelve todos los registros del DataFrame derecho (df_inventario) y agrega los de la izquierda si coinciden.
print(''' ðŸ‘‰ Se conservan todos los productos de df_inventario.
          ðŸ‘‰ E aparece con NaN en ventas porque no existe en df_concat.
          right: Todos de inventario (A, B, E) + coincidencias''')

df_combinado_right=pd.merge(df_concat, df_inventario, on="Producto", how="right")
display(df_combinado_right)
```

```python
# 4. outer (uniÃ³n completa)
# Devuelve todos los registros de ambos DataFrames.
# Si no hay coincidencia â†’ NaN.

print('''ðŸ‘‰ Se conservan todos los productos de ambos DataFrames
      ðŸ‘‰ Se conserva todo (A, B, C, D, E), aunque no coincidan en ambos.
      outer: UniÃ³n de todos (A, B, C, D, E)''')

df_combinado_outer=pd.merge(df_concat, df_inventario, on="Producto", how="outer")
display(df_combinado_outer)
```

ðŸ“Œ Resumen rÃ¡pido:
how	QuÃ© devuelve
inner	Solo coincidencias (intersecciÃ³n)
left	Todos los de la izquierda + coincidencias de la derecha
right	Todos los de la derecha + coincidencias de la izquierda
outer	Todos los registros de ambos (uniÃ³n completa)
