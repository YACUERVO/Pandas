```python
#Eliminar los datos duplicados en las filas
#Identifica los datos duplicados
duplicated_rows = df[df.duplicated()]
display(duplicated_rows)

```

```python
#Identificar duplicados por columna 
duplicated_by_name = df[df.duplicated(subset=['nombre'])]
display(duplicated_by_name)
```

```python
#Eliminar los datos duplicados
df_no_duplicated = df.drop_duplicates()
display(df_no_duplicated)
```

```python
#Tratamiendo de duplicados y mantener la aparicio nos sirve para quedarnos con la utlima actualizacion
#Para quedarse con el primero o el ultimo dato en la base de datos
df_last_ocurrence = df.drop_duplicates(subset=['nombre'],keep='last')
display(df_last_ocurrence)
```
