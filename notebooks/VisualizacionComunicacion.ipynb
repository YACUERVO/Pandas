```python
import pandas as pd
```

```python
#Como comunicar datos con visualizaciones

df = pd.read_csv('/Users/Admin/Documents/proyectosAnalisisDatos/Pandas/datos_con_faltantes.csv')

df.head()
```

```python
#Mapa de calor 
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Mapa de calor de los datos con faltantes')
plt.show()
```

```python
#Generar una matriz de datos faltantes
import missingno as msno
msno.matrix(df)
plt.title('Matrix de datos faltantes')
plt.show()
```

```python
#Grafico de barras por columna de datos faltantes
df.isnull().sum().plot(kind='bar')
plt.title('Grafico de barras por columna de datos faltantes')
plt.xlabel('Columnas')
plt.ylabel('Cantidad de datos faltantes')
plt.show()
```

```python
#Correlacion de los datos faltantes
msno.heatmap(df)
plt.title('Mapa de correlacion de los datos faltantes')
plt.show()
```
